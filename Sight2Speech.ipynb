```python
from PIL import Image
import pytesseract
from gtts import gTTS
import IPython.display as ipd

# Set up pytesseract to use default Tesseract
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

# Extracts text from the uploaded image using OCR.
def extract_text(image_path):
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img)
    return text.strip()

# Converts the text to speech using gTTS and plays it.
def text_to_speech(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    tts.save("output.mp3")
    return "output.mp3"

# Change this to your uploaded image name
image_name = "sample.jpg"

# Extract and speak
extracted_text = extract_text(image_name)
print("📝 Extracted Text:\n", extracted_text)

if extracted_text:
    audio_file = text_to_speech(extracted_text)
    print("\n🔊 Playing Audio:")
    ipd.display(ipd.Audio(audio_file))
else:
    print("⚠️ No text found.")
```
